# 6.1 Introduction

# 6.2 Design Decision: ArUco Markers

ArUco markers are a type of fiducial marker used in the field of computer vision and augmented reality. These markers are essentially black and white squares or grids with a unique pattern encoded within them. The patterns consist of a combination of black and white squares arranged in a specific configuration, which enables computer vision systems to easily detect and recognize them.

ArUco markers are designed to be easily detectable and identifiable by cameras, even under varying lighting conditions and viewing angles. This makes them highly suitable for applications such as camera calibration, pose estimation, and object tracking in robotics, augmented reality, and other computer vision-related tasks.

The name "ArUco" is derived from the combination of "Augmented Reality" and "University of Cordoba," where the markers were originally developed. These markers have gained popularity due to their simplicity, versatility, and robustness in various computer vision applications.

Additionally, ArUco markers can be customized to encode different information, such as identification numbers or orientation cues, by altering the arrangement of black and white squares within the marker's pattern. This flexibility allows for the creation of markers tailored to specific applications and requirements.

# 6.3 Camera Matrix

In robotics, the camera matrix, often denoted as K, represents the intrinsic parameters of a camera. These parameters define the relationship between the coordinates of points in a 3D scene and their corresponding coordinates in the image captured by the camera.

The camera matrix typically consists of:

*  Focal Lengths (fx and fy) These values represent the focal lengths of the camera's lens along the x and y axes, respectively. They determine how much the camera can magnify or reduce the size of objects in the image.
*   Principal Point (cx and cy): This denotes the coordinates of the principal point, which is the intersection of the image plane's optical axis with the image plane itself. It represents the location of the camera's optical center in the image.
*Skew (s): This parameter, often assumed to be zero in most cases, represents the skew between the x and y axes of the image plane.




image.png

**Figure: Camera Matrix**

Understanding the Camera Matrix Equation: The camera matrix equation encapsulates the intrinsic parameters of the camera, such as focal length, principal point, and skew. It relates the coordinates of a point in the 3D world (usually denoted as X) to its corresponding coordinates in the 2D image plane (denoted as x). Mathematically, this relationship is represented x=K⋅[R∣t]⋅X, where K is the camera matrix, R is the rotation matrix, t is the translation vector, and X and x represent the 3D world and 2D image coordinates, respectively.

Camera Calibration: By solving the camera matrix equation through camera calibration, we determine the intrinsic parameters of the camera, including the focal lengths, principal point, and skew. This calibration process involves capturing images of known calibration patterns, such as a chessboard, from different viewpoints. By detecting and matching corresponding points between the calibration pattern in the 3D world and its 2D projection in the image, we can compute the camera matrix K that best describes the camera's internal geometry.

Conversion of 2D to 3D Coordinates: Once we have the camera matrix K, along with the rotation matrix R and translation vector t, we can use the camera matrix equation to convert 2D image coordinates (e.g., pixel coordinates) into their corresponding 3D world coordinates. This process, known as perspective projection, enables us to reconstruct the 3D structure of objects or scenes captured by the camera based on their 2D projections in the image.

Applications: The ability to convert 2D image coordinates into 3D world coordinates is fundamental in various computer vision applications, such as 3D reconstruction, object tracking, pose estimation, and augmented reality. It allows us to extract depth information from 2D images and accurately model the geometry of the scene, enabling robots or other systems to perceive and interact with their environment in a more meaningful way.
